```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
from utils import load_mnist
from trainer import Trainer
```

```python
from argparse import Namespace

config = {
    'train_ratio': .8,
    'batch_size': 256,
    'n_epochs': 15,
    'verbose': 1,
    'btl_size': 2
}

config = Namespace(**config)

print(config)
```

```python
def show_image(x):
    if x.dim() == 1:
        x = x.view(int(x.size(0) ** .5), -1)
    
    plt.imshow(x, cmap='gray')
    plt.show()
```

```python
train_x, train_y = load_mnist(flatten=True)
test_x, test_y = load_mnist(is_train=False, flatten=True)

train_cnt = int(train_x.size(0) * config.train_ratio) # (batch_size, 784) == (60000, 784) 중 60000 * config.train_ratio
valid_cnt = train_x.size(0) - train_cnt

# Shuffle dataset to split into train/valid set.
indices = torch.randperm(train_x.size(0)) # 0~59999로 이루어지고, 순서는 랜덤인 텐서리스트를 만든다 == 인덱스를 섞어줌
train_x, valid_x = torch.index_select(
    train_x,
    dim=0, # 60000
    index=indices
).split([train_cnt, valid_cnt], dim=0)
train_y, valid_y = torch.index_select(
    train_y,
    dim=0, # 60000
    index=indices
).split([train_cnt, valid_cnt], dim=0)

print("Train:", train_x.shape, train_y.shape)
print("Valid:", valid_x.shape, valid_y.shape)
print("Test:", test_x.shape, test_y.shape)
```

```python
from model import Autoencoder
```

```python
model = Autoencoder(btl_size=config.btl_size)
optimizer = optim.Adam(model.parameters())
crit = nn.MSELoss()

trainer = Trainer(model, optimizer, crit)
```

- autoencoder를 학습시키는 것이라서
- 입력과 출력이 모두 x임
    - x -> z -> x_hat

```python
trainer.train((train_x, train_x), (valid_x, valid_x), config)
```

## Test

```python
with torch.no_grad():
    import random

    index = int(random.random() * test_x.size(0)) # 0~9999중 랜덤
    # test_x[index] == (784,)
    # test_x[index].view(1, -1) == (1, 784)
    recon = model(test_x[index].view(1, -1)).squeeze()

    show_image(test_x[index])
    show_image(recon)
```

## Latent Space

```python
if config.btl_size == 2:
    color_map = [
        'brown', 'red', 'orange', 'yellow', 'green',
        'blue', 'navy', 'purple', 'gray', 'black',
    ]
    
    plt.figure(figsize=(20, 10))
    with torch.no_grad():
        latents = model.encoder(test_x[:1000])
        
        for i in range(10):
            target_latents = latents[test_y[:1000] == i]
            target_y = test_y[:1000][test_y[:1000] == i]
            plt.scatter(target_latents[:, 0], ## x좌표
                        target_latents[:, 1], ## y좌표
                        marker='o',
                        color=color_map[i],
                        label=i)
        
        plt.legend()
        plt.grid(axis='both')
        plt.show()
```

```python
if config.btl_size == 2:
    min_range, max_range = -2., 2.
    n = 20
    step = (max_range - min_range) / float(n) # 0.2
    
    with torch.no_grad():
        lines = []

        for v1 in np.arange(min_range, max_range, step): # -2, -1.8, -1.6, ... , 1.8
            z = torch.stack([
                torch.FloatTensor([v1] * n), # [-2, -2, ... , -2]
                torch.FloatTensor([v2 for v2 in np.arange(min_range,
                                                          max_range, step)]),
            ], dim=-1) # (20, 2)
            # model.decoder(z).shape == (20, 784)
            # model.decoder(z).view(n, 28, 28) == (20, 28, 28)
            line = torch.clamp(model.decoder(z).view(n, 28, 28), 0, 1)
            line = torch.cat([line[i] for i in range(n - 1, 0, -1)], dim=0) # (28x20, 28) x 20
            lines += [line] # (28x20, 28x20)
            
        lines = torch.cat(lines, dim=-1)
        plt.figure(figsize=(20, 20))
        show_image(lines)
```
