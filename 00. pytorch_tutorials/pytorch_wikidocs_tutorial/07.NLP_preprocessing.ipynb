### 단어 집합(Vocabulary) 생성

```python
import urllib.request
import pandas as pd
from konlpy.tag import Okt
from nltk import FreqDist
import numpy as np
import matplotlib.pyplot as plt

urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings.txt", filename="ratings.txt")
data = pd.read_table('ratings.txt')
print(data.shape)
data.head()
```

```python
sample_data = data[:200] # 임의로 200개만 저장
sample_data['document'] = sample_data['document'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
sample_data.head()
```

```python
# 불용어 정의
stopwords=['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']
```

```python
okt = Okt()
def tokenize(sentence):
    words = okt.morphs(sentence)
    words = [word for word in words if word not in stopwords]
    return words

tokenized = []
for i in sample_data['document']:
    tokenized.append(tokenize(i))
print(tokenized[:5])
```

```python
vocab = FreqDist(np.hstack(tokenized))
print(vocab.elements)
print('단어 집합의 크기 : {}'.format(len(vocab)))
```

```python
vocab_size = 800
vocab = vocab.most_common(vocab_size)
len(vocab)
```

### 각 단어에 고유한 정수 부여

```python
word_to_index = {word[0] : idx+2 for idx, word in enumerate(vocab)}
word_to_index['pad'] = 1
word_to_index['unk'] = 0
word_to_index
```

```python
encoded = []
for words in tokenized:
    tmp = []
    for word in words:
        try:
            tmp.append(word_to_index[word])

        except KeyError:
            tmp.append(word_to_index['unk'])
    encoded.append(tmp)
print(encoded[:5])
```

### 길이가 다른 문장들을 모두 동일한 길이로 바꿔주는 패딩(padding)

```python
max_len = max(len(line) for line in encoded)
min_len = min(len(line) for line in encoded)
mean_len = sum(map(len, encoded)) / len(encoded)
print ('최대길이: {}, 최소길이: {}, 평균길이: {}'.
       format(max_len, min_len, mean_len))

plt.hist([len(s) for s in encoded], bins=50)
plt.xlabel('length of sample')
plt.ylabel('number of sample')
plt.show()
```

```python
word_to_index['pad']
```

```python
for idx, line in enumerate(encoded):
    if len(line) < max_len:
        for i in range(max_len - len(line)):
            encoded[idx].append(word_to_index['pad'])
            
mean_len = sum(map(len, encoded)) / len(encoded)
mean_len
```

### 토치텍스트 튜토리얼(Torchtext tutorial) - 영어
- 파일 로드하기(File Loading) : 다양한 포맷의 코퍼스를 로드합니다.
- 토큰화(Tokenization) : 문장을 단어 단위로 분리해줍니다.
- 단어 집합(Vocab) : 단어 집합을 만듭니다.
- 정수 인코딩(Integer encoding) : 전체 코퍼스의 단어들을 각각의 고유한 정수로 맵핑합니다.
- 단어 벡터(Word Vector) : 단어 집합의 단어들에 고유한 임베딩 벡터를 만들어줍니다. 랜덤값으로 초기화한 값일 수도 있고, 사전 훈련된 임베딩 벡터들을 로드할 수도 있습니다.
- 배치화(Batching) : 훈련 샘플들의 배치를 만들어줍니다. 이 과정에서 패딩 작업(Padding)도 이루어집니다.

#### 5. model 만들기

- BucketIterator를 거친 후의 결과물은, 각 단어를 숫자에 매핑하여 배치로 나누어준것뿐..
- 이후에 따로 embedding을 거쳐서 단어간에 연관성을 주어야함!!
- 그래서 embedding layer의 input으로 len(TEXT.vocab)을 넣어줌

```python
class DNN(nn.Module):
    def __init__(self, input_size, embedded_size, hidden_size, n_classes, n_layers=4, dropout_p=0.3):
        self.input_size = input_size
        self.embedded_size = embedded_size
        self.n_classes = n_classes
        super().__init__()
        
        self.emb_layer = nn.Embedding(input_size, embedded_size)
        self.fc_layer = nn.Linear(embedded_size, 64)
        self.relu = nn.ReLU()
        self.fc_layer2 = nn.Linear(64, n_classes)
        self.activation = nn.LogSoftmax(dim=-1)
        
    def forward(self, x):
#         print((x.shape))
        z = self.emb_layer(x)
#         print((z.shape))
        z = self.relu(self.fc_layer(z))
        y = self.fc_layer2(z[:, -1])
#         print(y.shape)
        return self.activation(y)
```

```python
class LSTM_model(nn.Module):
    def __init__(self, input_size, embedded_size, hidden_size, n_classes, n_layers=2, dropout_p=0.3):
        self.input_size = input_size
        self.embedded_size = embedded_size
        self.hidden_size = hidden_size
        self.n_classes = n_classes
        self.n_layers = n_layers
        self.dropout_p = dropout_p
        super().__init__()
        
        self.emb = nn.Embedding(input_size, embedded_size)
        self.lstm = nn.LSTM(
            input_size = embedded_size,
            hidden_size = hidden_size,
            num_layers = n_layers,
            dropout = dropout_p,
            batch_first = True,
            bidirectional = True,
        )
        self.generator = nn.Linear(hidden_size * 2, n_classes)
        self.activation = nn.LogSoftmax(dim=-1)
        
    def forward(self, x):
        x = self.emb(x)
        # |x| = (batch_size, length, word_vec_size)
        x, _ = self.lstm(x)
        # |x| = (batch_size, length, hidden_size * 2)
        y = self.activation(self.generator(x[:, -1]))
        return y
```

```python
from easydict import EasyDict
from tqdm import tqdm
from copy import deepcopy

config = EasyDict({'input_size' : len(TEXT.vocab),
                   'embedded_size' : 64,
                   'hidden_size': 128,
                   'n_classes' : 2,
                   'EPOCHS' : 3})

lstm_model = LSTM_model(config.input_size, config.embedded_size, config.hidden_size, config.n_classes).to('cuda:0')
optimizer = optim.Adam(lstm_model.parameters())
# crit = nn.CrossEntropyLoss().to('cuda:0')
# crit = nn.BCELoss().to('cuda:0')
crit = nn.NLLLoss().to('cuda:0')
best_loss = 999999
best_model = None

for i in range(config.EPOCHS):
    train_loss = 0
    for x, y in tqdm(train_loader):
        lstm_model.train()
        optimizer.zero_grad()
        
        x, y = x.to('cuda:0'), y.to('cuda:0')
        x = x[:, :1000]
        y_hat =lstm_model(x)
        loss = crit(y_hat, y)
        loss.backward()
        optimizer.step()
        train_loss += float(loss) / len(train_loader) # loss / 미니배치 수
    
    valid_loss = 0
    for x, y in valid_loader:
        lstm_model.eval()
        with torch.no_grad():
            x, y = x.to('cuda:0'), y.to('cuda:0')
            x = x[:, :1000]
            y_hat = lstm_model(x)
            loss = crit(y_hat, y)
            
            if loss < best_loss:
                best_loss = loss
                best_model = deepcopy(lstm_model.state_dict())
        
            valid_loss += float(loss) / len(valid_loader)
    print('EPOCHS: {} | train_loss: {:.5f} valid_loss: {:.9f}'.format(i+1, train_loss, valid_loss))
        
torch.save({
        'lstm': lstm_model.state_dict(),
        'config': config,
        'vocab': TEXT.vocab,
        'num_classes' : 2
    }, 'lstm_model.pth')
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.functional as F
```

```python
batch_size = 16
embedded_size = 64
hidden_size = 128
n_classes = 2
EPOCHS = 5
```

```python
class LSTM(nn.Module):
    def __init__(self, 
                 input_size, 
                 embedded_size, 
                 hidden_size, 
                 n_classes, 
                 n_layers=3, 
                 dropout_p=0.3):
        self.input_size = input_size
        self.embedded_size = embedded_size
        self.hidden_size = hidden_size
        self.n_classes = n_classes
        self.n_layers = n_layers
        self.dropout_p = dropout_p
        super().__init__()
        
        # 단순히 숫자로 매칭만 되어있던 것을 neural net에 넣어서 학습시켜줌
        self.emb = nn.Embedding(self.input_size, self.embedded_size) 
        self.lstm = nn.LSTM(
            input_size = self.embedded_size,
            hidden_size = self.hidden_size,
            num_layers = self.n_layers,
            dropout = self.dropout_p,
            batch_first = True,
            bidirectional = True
        )
        self.fc_layer = nn.Linear(hidden_size*2, n_classes)
        self.activation = nn.LogSoftmax(dim=-1)
        
    def forward(self, x):
        # |x| = (bs, length)
        
        embedded = self.emb(x) 
        # [embedded] = (bs, length, embedded_size)
        
        y, _ = self.lstm(embedded)
        # |y| = (bs, length, hidden_size*2)

        y = self.fc_layer(y[:, -1, :]) # lstm의 결과중 마지막것만 가져옴
        # [y] = (bs, length, n_classes)
        
        y = self.activation(y)
        # [y] = (bs, length, n_classes)
        return y
```

```python
cnt = 0
for x, y in train_loader:
    if cnt == 0:
        print(x[0]) # idx... 여기 필요없음
        print(x[1]) # 이부분만 가져올것
    cnt += 1
```

```python
from tqdm import tqdm

model = LSTM(len(TEXT.vocab), embedded_size, hidden_size, n_classes).to('cuda:0')
crit = nn.NLLLoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)
best_loss = 999999
best_model_kor = None

for i in range(EPOCHS):
    train_loss = 0
    train_accuracy = 0
    model.train()
    for x, y in tqdm(train_loader):
        optimizer.zero_grad()
        x = x[1]
        x, y = x.to('cuda:0'), y.to('cuda:0')

        y_hat = model(x)
        loss = crit(y_hat, y)
        loss.backward()
        optimizer.step()
        if isinstance(y, torch.LongTensor) or isinstance(y, torch.cuda.LongTensor):
            accuracy = (torch.argmax(y_hat, dim=-1) == y).sum() / float(y.size(0))
        else:
            accuracy = 0
        train_loss += float(loss) / len(train_loader)
        train_accuracy += accuracy / len(train_loader)
    
    valid_loss = 0
    valid_accuracy = 0
    with torch.no_grad():
        model.eval()
        for x, y in valid_loader:
            x = x[1]
            x, y = x.to('cuda:0'), y.to('cuda:0')
            y_hat = model(x)
            loss = crit(y_hat, y)
            
            if isinstance(y, torch.LongTensor) or isinstance(y, torch.cuda.LongTensor):
                accuracy = (torch.argmax(y_hat, dim=-1) == y).sum() / float(y.size(0))
            else:
                accuracy = 0
            valid_loss += float(loss) / len(valid_loader)
            valid_accuracy += accuracy / len(valid_loader)
            
            if loss < best_loss:
                best_loss = loss
                best_model_kor = deepcopy(model.state_dict())
            
            
    print("EPOCHS: {:2d} | train_accuracy: {:.4f} / train_loss: {:.4f} / valid_accuracy: {:.4f} / valid_loss: {:.4f}".format
              (i+1, train_accuracy, train_loss, valid_accuracy, valid_loss))
```
